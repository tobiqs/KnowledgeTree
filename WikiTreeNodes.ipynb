```python
import requests
import bs4 as bs
import pandas as pd
import hashlib
```

```python
class Node:
    "Apple of the KnowledgeTree :)"
    
    
    url_wiki = "https://en.wikipedia.org"
    
    def __init__(self, title = 'Algebra', link = '/wiki/Algebra'):
        self.title = title
        self.link = link
        self.branches = []
        
        #generate hash (MD5)
        
        hash_obj = hashlib.md5(link.encode())
        h = hash_obj.hexdigest()     
        
        self.hash = h
        
    def info(self):
        print("node info")
        print("title: %s | link: %s "%(self.title, self.link))
        print("branches: %i"%(len(self.branches)))

    def grabPage(self):
        url = self.link
        r = requests.get(Node.url_wiki + url)
        self.rawPage = r.content
        self.soupPage = bs.BeautifulSoup(self.rawPage, 'lxml')
        
    def actualizeNode(self):
        "creating branch nodes"
        print("Actualize node: creating branches")
        
        s = self.soupPage
        
        for a in s.find_all('a'):

            if (a.get('href')!=None):
                if (a.get('title')!=None):
            
                    link = a['href']
                    title = a['title']
                    newNode = Node(title = title, link = link)
                    
                    self.branches.append(newNode)
        print(' %i nodes created. '%len(self.branches))

    def createNodeDF(self):
        print("creating DataFrame of nodes' branches")
        self.df_nodes = pd.DataFrame()
        
        hashes = []
        titles = []
        links = []
        
        for bran in self.branches:
            hashes.append(bran.hash)
            titles.append(bran.title)
            links.append(bran.link)
            
            
        self.df = pd.DataFrame({'titles': titles, 'links': links, 'hashes': hashes})
        
    def pageStats(self):
        print(" #links on page: %i "%len(self.soupPage.find_all('a')))
        
        
    def get(self):

        self.grabPage()
        self.info()

    def update(self):

        # how many potiential branches

        self.pageStats()

        # parse data

        self.actualizeNode()
        self.createNodeDF()

        self.info()      
        
    def save(self):
        
        # please create 'data' directory if not existent
        
        self.filename = 'data/' + self.hash
        self.df.to_csv(self.filename + '.csv')
        self.df.to_json(self.filename + '.json')

```

```python
ROOT = Node()

ROOT.get()
ROOT.update()
ROOT.save()
```
